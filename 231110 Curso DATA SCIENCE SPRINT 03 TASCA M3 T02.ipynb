```python
# 3.- Eliminar las columnas que no me interesan: 
# Con drop(), e indicándole que es el eje 1, el de las columnas:

columnas_a_eliminar = ['Year','Month','DayofMonth', 'DepTime','CRSDepTime',
                       'ArrTime','CRSArrTime','DepDelay','Cancelled','Diverted']
Delay_Table_3 = Delay_Table_2.drop(columnas_a_eliminar, axis=1)

num_columnas_3 = Delay_Table_3.shape[1]
print ('la tabla Delay_Table_3 ha quedado constituida por un número final de:', num_columnas_3,'columnas')
```

```python
# 4.- Reducir la dimensión de la tabla de manera aleatoria dejando solo 200.000 registros

filas_final = 200000  

# Obtengo una muestra aleatoria de n filas con sample()
filas_a_mantener = Delay_Table_3.sample(filas_final) 

# Elimino las filas que no están en la muestra aleatoria con drop () y difrerence()
# Con la extensión .index me aseguro que que la tabla final solo contenga las filas que están en la muestra aleatoria
Delay_Final = Delay_Table_3.drop(Delay_Table_3.index.difference(filas_a_mantener.index))

num_filas_Final = Delay_Final.shape[0]
print ('la nueva tabla Delay_Final ha quedado compuesta compuesta por:', num_filas_Final,'filas')
```
